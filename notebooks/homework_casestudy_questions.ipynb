# Easy Level (Questions 1–8)

These questions can be answered using simple queries or basic aggregation.

---

## Question 1

### Question

**What is the total number of sales transactions recorded in the dataset?**

### Business Objective

Management wants to determine the total number of sales transactions recorded by the business. This serves as a baseline Key Performance Indicator (KPI) for understanding the overall business activity.
